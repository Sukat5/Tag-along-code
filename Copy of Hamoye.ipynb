```python
import pandas as pd
import numpy as np
```

```python
csv_df= pd.read_csv('/content/sample_data/FoodBalanceSheets_E_Africa_NOFLAG.csv', encoding='latin-1')
```

```python
#Q2
selected_columns=csv_df[['Y2017', 'Area']]
grouped= csv_df.groupby('Area')
sum_by_area = grouped['Y2017'].sum()
area_with_highest_sum = sum_by_area.idxmax()
print("Area with the highest sum in 2017:", area_with_highest_sum)
```

```python
#Q5
df_2017 = csv_df['Y2017']
mean_2017 = df_2017.mean()
std_2017 = df_2017.std()
print("Mean for the year 2017:", round(mean_2017, 2))
print("Standard deviation for the year 2017:", round(std_2017, 2))
```

```python
#Q6
columns_to_analyze = ['Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018','Element Code']
subset_df = csv_df[columns_to_analyze]

# Calculating the correlation between each year column and 'Element Code'
correlations = subset_df.iloc[:, :-1].apply(lambda col: col.corr(subset_df['Element Code']))

# Finding the year with the least correlation
year_with_least_correlation = correlations.idxmin()

print("Year with the least correlation with 'Element Code':", year_with_least_correlation)
```

```python
#Q7
grouped = csv_df.groupby('Element')
```

```python
#Q7
filtered_df = csv_df[csv_df['Element'] == 'Stock Variation']
sum_by_year = filtered_df[['Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018']].sum()
highest_sum_year = sum_by_year.idxmax()
print("Year with the highest sum of Stock Variation:", highest_sum_year)
```

```python
#Q8
filtered_df = csv_df[(csv_df['Element'] == 'Protein supply quantity') & (csv_df['Area'] == 'Madagascar')]

# Retrieving the Protein supply quantity for the year 2015
protein_supply_2015 = filtered_df['Y2015'].sum()

print("Total Protein supply quantity in Madagascar in 2015:", protein_supply_2015)
```

```python
#Q9
wine_rows = csv_df[csv_df['Item'] == 'Wine']
year_columns = ['Y2015', 'Y2018']

# Calculating the total sum of Wine produced in 2015 and 2018
wine_sum_2015 = wine_rows[year_columns[0]].sum()
wine_sum_2018 = wine_rows[year_columns[1]].sum()

print("Total sum of Wine produced in 2015:", wine_sum_2015)
print("Total sum of Wine produced in 2018:", wine_sum_2018)
```

```python
#Q10
selected_columns = ['Y2017', 'Area']
df_selected = csv_df[selected_columns]

# Perform groupby and calculate the sum for each area in 2017
grouped = df_selected.groupby('Area').sum()

# Sort the sums in ascending order
sorted_sums = grouped.sort_values('Y2017')

# Find the area with the 7th lowest sum in 2017
seventh_lowest_area = sorted_sums.index[6]

print("Area with the 7th lowest sum in 2017:", seventh_lowest_area)
```

```python
#Q14
year_2014_column = csv_df['Y2014']
missing_data = year_2014_column.isnull().sum()

# percentage of missing data
total_count = len(year_2014_column)
percentage_missing = (missing_data / total_count) * 100

#results
print("Total number of missing values in 2014:", missing_data)
print("Percentage of missing data in 2014:", round(percentage_missing, 3))
```

```python
#Q15
filtered_df = csv_df[csv_df['Element'] == 'Processing']

# Select the columns for the year 2017 (Y2017)
columns_2017 = ['Y2017']
df_2017 = filtered_df[columns_2017]

# Calculate the sum of 'Processing' in 2017
processing_sum_2017 = df_2017.sum().sum()

print("Total sum of Processing in 2017:", processing_sum_2017)
```
